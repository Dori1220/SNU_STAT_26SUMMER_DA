```python
from pathlib import Path
import pandas as pd
```

```python
df = pd.read_csv('original_data/소상공인시장진흥공단_상가(상권)정보_충남_202603.csv', dtype={"층정보": str}, encoding='utf-8')
```

```python
df.columns
```

```python
df_cheonan = df[df["시군구명"].astype(str).str.contains("천안", na=False)]
```

```python
df_cheonan["시군구명"].value_counts()
```

```python
out_path = Path("./commerce_cheonan_202603.csv")
```

```python
df_cheonan.to_csv(
    out_path,
    index=False,
    encoding="utf-8-sig"
)

print("Done.")
print(f"Cheonan rows: {len(df_cheonan):,}")
print(f"Saved to: {out_path.resolve()}")
```
