```python
from pathlib import Path
import pandas as pd


# =========================
# 1. Settings
# =========================

files = [
    {
        "in_path": Path("./original_data/1.병원정보서비스(2026.3.).xlsx"),
        "out_path": Path("./hospital_cheonan_202603.csv")
    },
    {
        "in_path": Path("./original_data/2.약국정보서비스(2026.3.).xlsx"),
        "out_path": Path("./pharmacy_cheonan_202603.csv")
    }
]

target_col = "시군구코드명"
keyword = "천안"


# =========================
# 2. Convert xlsx to filtered csv
# =========================

for file in files:
    in_path = file["in_path"]
    out_path = file["out_path"]

    df = pd.read_excel(
        in_path,
        sheet_name=0,
        dtype=str,
        engine="openpyxl"
    )

    # 컬럼명 앞뒤 공백 제거
    df.columns = df.columns.astype(str).str.strip()

    if target_col not in df.columns:
        raise ValueError(
            f"{in_path.name}: '{target_col}' 컬럼이 없습니다. "
            f"실제 컬럼명: {df.columns.tolist()}"
        )

    cheonan = df[
        df[target_col]
        .astype(str)
        .str.contains(keyword, na=False)
    ].copy()

    cheonan.to_csv(
        out_path,
        index=False,
        encoding="utf-8-sig"
    )

    print("Done:", in_path.name)
    print(f"Original rows: {len(df):,}")
    print(f"Cheonan rows: {len(cheonan):,}")
    print(f"Saved to: {out_path.resolve()}")
    print("-" * 60)
```
