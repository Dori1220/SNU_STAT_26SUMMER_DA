# 이론적 power 파생변수 계산 (일단 group1 대상) - LDAPS 버전

### 1. Group 선택 (1~3)  
### 2. 선택된 group에 소속된 터빈의 위경도, 로터 직경, 허브 높이 읽어오기  
### LDAPS data에 포함된 grid_id별 위경도 읽어오기
### 3. 각 터빈 위치에서 가장 가까운 grid_id 대응 (k개)

```python
def get_nearest_ldaps_grid(
        kpx_group: int,
        k: int,
        info_path: str = "../data/info_include_lat_lon.csv",
        ldaps_path: str = ""
) -> pd.DataFrame:
    
    # 1. info 로드 및 group 필터링
    info = pd.read_csv(info_path, encoding = 'utf-8-sig')
    group_info = info[info['KPX그룹'] == kpx_group].copy()

    # 2. 선택된 group에 소속된 터빈의 위경도, 로터 직경, 허브 높이
    turbines = group_info[[
        '제작사', '모델명', '호기', 'latitude', 'longitude', 'Hub Height(m)', 'Rotor Diameter(m)'
        ]].reset_index(drop = True)
    
    # 3. LDAPS grid_id별 위경도
    ldaps = pd.read_csv(ldaps_path, encoding = 'utf-8-sig', 
                        usecols = ['grid_id', 'latitude', 'longitude'])
    grids = (
        ldaps.drop_duplicates('grid_id')
        .sort_values('grid_id')
        .reset_index(drop = True)
    )

    # 4. 각 터빈 위치에서 가장 가까운 grid_id k개를 haversine 거리로 대응
    def haversine_km(lat1, lon1, lat2, lon2):
        R = 6371.0088
        lat1, lon1, lat2, lon2 = map(np.radians, (lat1, lon1, lat2, lon2))
        dlat = lat2 - lat1
        dlon = lon2 - lon1
        a = np.sin(dlat / 2) ** 2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2) ** 2
        return 2 * R * np.arcsin(np.sqrt(a))
    
    grid_lat = grids['latitude'].to_numpy()
    grid_lon = grids['longitude'].to_numpy()
    grid_ids = grids['grid_id'].to_numpy()

    records = []
    for _, t in turbines.iterrows():
        dist = haversine_km(t['latitude'], t['longitude'], grid_lat, grid_lon)
        nearest_idx = np.argsort(dist)[:k]
        for rank, gi in enumerate(nearest_idx, start = 1):
            records.append({
                '제작사': t['제작사'],
                '모델명': t['모델명'],
                '호기': t['호기'],
                'turbine_lat': t['latitude'],
                'turbine_lon': t['longitude'],
                'Hub Height(m)': t['Hub Height(m)'],
                'Rotor Diameter(m)': t['Rotor Diameter(m)'],
                'rank': rank,
                'grid_id': grid_ids[gi],
                'grid_lat': grid_lat[gi],
                'grid_lon': grid_lon[gi],
                'dist_km': dist[gi]
            })
    
    return pd.DataFrame(records)
```

### 4. 허브높이 H에서의 풍속으로 보정 (grid 별로)  

```python
ldaps_train_group1 = pd.read_csv("../data/ldaps_train_group1.csv", encoding = 'utf-8-sig')
```

```python
# group1 nearest info ldaps 생성
nearest_info_ldaps_group1 = get_nearest_ldaps_grid(kpx_group=1, k=4, ldaps_path='../data/ldaps_train_group1.csv')[
    ['제작사', '호기', 'Hub Height(m)', 'Rotor Diameter(m)', 'rank', 'grid_id', 'dist_km']
    ]
nearest_info_ldaps_group1
```

```python
# 허브높이, 기준 높이 변수 생성
HUB_HEIGHT = 117
REF_LOW = 10
REF_HIGH = 50
```

```python
# 10m 풍속 feature 생성
ldaps_train_group1['ws10_ldaps'] = np.hypot(
    ldaps_train_group1['heightAboveGround_10_10u'],
    ldaps_train_group1['heightAboveGround_10_10v']
)

# 50m 풍속 feature 생성
u50_ldaps = (
    ldaps_train_group1['heightAboveGround_50_50MUmax'] + 
    ldaps_train_group1['heightAboveGround_50_50MUmin']
) / 2
v50_ldaps = (
    ldaps_train_group1['heightAboveGround_50_50MVmax'] + 
    ldaps_train_group1['heightAboveGround_50_50MVmin']
) / 2
ldaps_train_group1['ws50_ldaps'] = np.hypot(u50_ldaps, v50_ldaps)

# 풍속 전단 지수 alpha feature 생성
ws10_ldaps_safe = np.clip(ldaps_train_group1['ws10_ldaps'], 0.5, None) # 저풍속 구간 log 발산 방지
ws50_ldaps_safe = np.clip(ldaps_train_group1['ws50_ldaps'], 0.5, None) # 저풍속 구간 log 발산 방지
ldaps_train_group1['alpha'] = np.log(ws50_ldaps_safe / ws10_ldaps_safe) / np.log(REF_HIGH / REF_LOW)
ldaps_train_group1['alpha'] = ldaps_train_group1['alpha'].clip(0.05, 0.4) # 물리적으로 타당한 범위로 제한

# 허브높이 풍속 feature 생성
ldaps_train_group1['ws_hub_ldaps'] = ldaps_train_group1['ws50_ldaps'] * (HUB_HEIGHT / REF_HIGH) ** ldaps_train_group1['alpha']
```

### 5. 허브높이 H에서의 기온으로 보정 (grid 별로)  

```python
# 표준 기온감률, 2m 기준고도 변수 정의
LAPSE_RATE = 0.0065
REF_T = 2
```

```python
# 허브높이 기온 feature 생성
ldaps_train_group1['t_hub_ldaps'] = ldaps_train_group1['heightAboveGround_2_t'] - LAPSE_RATE * (HUB_HEIGHT - REF_T)
```

### 6. 허브높이 H에서의 공기밀도 계산 (grid 별로)

```python
# 중력가속도, 건조공기 기체상수 변수 정의
G = 9.80665
R_D = 287.05
```

```python
# 가상온도 계산
q = ldaps_train_group1['heightAboveGround_2_q']
tv_sfc = ldaps_train_group1['heightAboveGround_2_t'] * (1 + 0.61 * q) # 지표 부근 가상온도
tv_hub = ldaps_train_group1['t_hub_ldaps'] * (1 + 0.61 * q) # 허브높이 가상온도

# 허브높이 기압 계산
tv_mean = (tv_sfc + tv_hub) / 2
p_hub = ldaps_train_group1['surface_0_sp'] * np.exp(-G * HUB_HEIGHT / (R_D * tv_mean))

# 허브높이 공기밀도 feature 생성
ldaps_train_group1['rho_hub_ldaps'] = p_hub / (R_D * tv_hub)
```

### 7. 터빈별 이론적 power proxy 계산

```python
# 격자별 이론적 power proxy -> ldaps_train_group1에 추가
ROTOR_D = 126
SWEPT_AREA = np.pi * (ROTOR_D / 2) ** 2
CP = 0.35

ldaps_train_group1['power_proxy_ldaps'] = 0.5 * ldaps_train_group1['rho_hub_ldaps'] * SWEPT_AREA * ldaps_train_group1['ws_hub_ldaps'] ** 3 * CP / 1000

# 나중에 클리핑 필요하면 하기 (일단은 X)
```

```python
### 터빈별 이론적 power proxy 계산

# 터빈별 가중치
weight_ldaps_group1 = nearest_info_ldaps_group1[['호기', 'grid_id', 'dist_km']].copy()
weight_ldaps_group1['weight'] = 1 / weight_ldaps_group1['dist_km']
weight_ldaps_group1['weight'] = (
    weight_ldaps_group1['weight'] /
    weight_ldaps_group1.groupby('호기')['weight'].transform('sum')
)
```

```python
# LDAPS 데이터에 터빈-격자 가중치 결합
merged_ldaps_group1 = ldaps_train_group1[['forecast_kst_dtm', 'grid_id', 'power_proxy_ldaps']].merge(
    weight_ldaps_group1[['호기', 'grid_id', 'weight']],
    on = 'grid_id',
    how = 'inner'
)
```

```python
merged_ldaps_group1
```

```python
# 터빈별 거리 가중 평균
merged_ldaps_group1['power_proxy_ldaps'] = merged_ldaps_group1['power_proxy_ldaps'] * merged_ldaps_group1['weight']

turbine_proxy_ldaps_group1 = (
    merged_ldaps_group1
    .groupby(['forecast_kst_dtm', '호기'])['power_proxy_ldaps']
    .sum()
    .unstack('호기')
)
```

```python
turbine_proxy_ldaps_group1.columns = [
    f'turbine{호기}_power_proxy_ldaps' for 호기 in turbine_proxy_ldaps_group1.columns
]
turbine_proxy_ldaps_group1 = turbine_proxy_ldaps_group1.reset_index()
```

# LDAPS : 1시간 단위로 grid_id에 따른 대표값 설정

```python
# 1. 격자별 난류강도 파생변수 (50m max-min 폭)
ldaps_train_group1['gust50_ldaps'] = np.hypot(
    ldaps_train_group1['heightAboveGround_50_50MUmax'] - 
    ldaps_train_group1['heightAboveGround_50_50MUmin'], 
    ldaps_train_group1['heightAboveGround_50_50MVmax'] -
    ldaps_train_group1['heightAboveGround_50_50MVmin']
)
```

```python
# 2. 집계 대상 변수 (삭제 대상은 여기 미포함)
IDW_COLS = [
    'power_proxy_ldaps', 
    'ws_hub_ldaps', 'ws10_ldaps', 'ws50_ldaps', 'alpha', 'gust50_ldaps', 
    'heightAboveGround_10_10u', 
    'heightAboveGround_10_10v', 
    'rho_hub_ldaps', 't_hub_ldaps',
    'heightAboveGround_2_r', 'heightAboveGround_2_q',
    'surface_0_sp', 'etc_0_blh',
    'surface_0_NDNSW', 'surface_0_NDNLW',
    'etc_0_hcc', 'etc_0_mcc', 'etc_0_lcc',
    'surface_0_lssrate', 'surface_0_SNOM',
    'heightAboveGround_5_XBLWS', 'heightAboveGround_5_YBLWS'
]
```

```python
# 3. 터빈별 IDW 가중치
weight_ldaps_group1 = nearest_info_ldaps_group1[['호기', 'grid_id', 'dist_km']].copy()
weight_ldaps_group1['weight'] = 1 / weight_ldaps_group1['dist_km']
weight_ldaps_group1['weight'] = (
    weight_ldaps_group1['weight'] / weight_ldaps_group1.groupby('호기')['weight'].transform('sum')
)
```

```python
# 4. 터빈 단위로 축약
merged_ldaps_group1 = ldaps_train_group1[['forecast_kst_dtm', 'grid_id'] + IDW_COLS].merge(
    weight_ldaps_group1[['호기', 'grid_id', 'weight']], 
    on = 'grid_id', 
    how = 'inner'
)

for col in IDW_COLS:
    merged_ldaps_group1[col] = merged_ldaps_group1[col] * merged_ldaps_group1['weight']

turbine_level_ldaps_group1 = (
    merged_ldaps_group1
    .groupby(['forecast_kst_dtm', '호기'])[IDW_COLS]
    .sum()
)
```

```python
# 5. 그룹 단위로 축약 (시각당 1행)
MEAN_COLS = [c for c in IDW_COLS if c != 'power_proxy_ldaps']
agg_spec = {c: 'mean' for c in MEAN_COLS}
agg_spec['power_proxy_ldaps'] = 'sum' # label = 터빈 발전량의 합

grouped = turbine_level_ldaps_group1.groupby('forecast_kst_dtm')
ldaps_group1_hourly = grouped.agg(agg_spec)
ldaps_group1_hourly['ws_hub_std_ldaps'] = grouped['ws_hub_ldaps'].std() # 터빈 간 이질성

ldaps_group1_hourly = ldaps_group1_hourly.rename(
    columns = {'power_proxy_ldaps': 'group1_power_proxy_ldaps'}
)
```

```python
# 6. 풍향 : 성분 평균 후 각도 계산 (sin/cos 인코딩)
wd_rad = np.arctan2(-ldaps_group1_hourly['heightAboveGround_10_10u'], 
                    -ldaps_group1_hourly['heightAboveGround_10_10v'])

ldaps_group1_hourly['wd10_sin_ldaps'] = np.sin(wd_rad)
ldaps_group1_hourly['wd10_cos_ldaps'] = np.cos(wd_rad)
```

```python
# 7. 누수 검증용 시각 결합
avail_ldaps_group1 = ldaps_train_group1.groupby('forecast_kst_dtm')['data_available_kst_dtm'].first()
ldaps_group1_hourly = ldaps_group1_hourly.join(avail_ldaps_group1).reset_index()
```

```python
# 8. 시간 파생변수
dtm = pd.to_datetime(ldaps_group1_hourly['forecast_kst_dtm'])
ldaps_group1_hourly['hour']  = dtm.dt.hour     # 안정도 편향 + 예보 리드타임을 함께 담음
ldaps_group1_hourly['month'] = dtm.dt.month    # 계절별 예보 편향
```
